**Objective:** To evaluate the effectiveness of two distinct website landing page designs (Control vs. Variant) and determine which drove a higher user conversion rate using statistical hypothesis testing.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Set a professional style for business charts
sns.set_palette(['#3498db', '#e74c3c']) # Blue for Control, Red for Variant
sns.set_style('whitegrid')

# Load the dataset
# NOTE: Make sure you have downloaded an A/B testing CSV from Kaggle and named it 'ab_test_data.csv'
try:
    df = pd.read_csv('ab_test_data.csv')
    print("Data loaded successfully! Here are the first 5 rows:")
    display(df.head())
except FileNotFoundError:
    print("Error: Could not find 'ab_test_data.csv'. Please make sure the dataset is downloaded.")
```

## 1. Exploratory Data Analysis (EDA)
Before running statistical tests, let's look at the basic conversion rates for both the Control group (the old webpage) and the Variant group (the new webpage).

```python
if 'df' in locals() and 'group' in df.columns and 'converted' in df.columns:
    # Calculate basic conversion rates
    conversion_rates = df.groupby('group')['converted'].mean() * 100
    print("Conversion Rates (%):")
    print(conversion_rates)
    
    # Visualize the difference
    plt.figure(figsize=(8, 5))
    sns.barplot(x=conversion_rates.index, y=conversion_rates.values)
    plt.title('Conversion Rate by Group', fontsize=16, fontweight='bold')
    plt.xlabel('Test Group')
    plt.ylabel('Conversion Rate (%)')
    plt.ylim(0, max(conversion_rates.values) + 5)
    plt.show()
else:
    print("Waiting for data, or the dataset does not have 'group' and 'converted' columns.")
```

## 2. Statistical Hypothesis Testing (Two-Sample T-Test)
The Variant group might have a higher conversion rate, but is that difference statistically significant, or did it just happen by random chance? 

*   **Null Hypothesis (H0):** There is NO difference in conversion rates between the groups.
*   **Alternative Hypothesis (H1):** There IS a difference in conversion rates.

```python
if 'df' in locals() and 'group' in df.columns and 'converted' in df.columns:
    # Separate the data into two arrays
    control_results = df[df['group'] == 'control']['converted']
    variant_results = df[df['group'] == 'variant']['converted']
    
    # Perform the T-Test
    t_stat, p_value = stats.ttest_ind(variant_results, control_results)
    
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")
    print("-" * 30)
    
    # Interpret the results (using a standard 0.05 significance level)
    if p_value < 0.05:
        print("Conclusion: The p-value is less than 0.05. We REJECT the null hypothesis.")
        print("Business Impact: The new variant page caused a statistically significant increase in conversions. We should deploy the new design.")
    else:
        print("Conclusion: The p-value is greater than 0.05. We FAIL TO REJECT the null hypothesis.")
        print("Business Impact: The new variant page did not perform significantly better. We should stick with the old design and test a different idea.")
```
